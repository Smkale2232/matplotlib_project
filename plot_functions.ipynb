```python
import matplotlib.pyplot as plt
```

```python
months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]
plt.plot(months, sales, color='blue', linestyle= '--', linewidth= 2, marker= 'o', label= '2025 Sales Data' )
plt.title('Monthly Sales Data Report')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4'])
plt.show()
```

```python
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.barh(product, sales, color= 'orange', label= 'Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```

```python
region = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=region, autopct='%1.1f%%', colors=['gold', 'silver', 'skyblue', 'coral'])
plt.title('Revenue Contributuion')
plt.show()
```

```python
scores = [89, 54, 62, 84, 76, 96, 70, 84, 99, 57, 66, 88, 91, 75, 83, 76, 85, 92, 90]

plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')
plt.show()
```

```python
hours_studies = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studies, exam_scores, color='green', marker='o', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel( 'Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
