```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,30]
plt.plot(x,y)
plt.show()
```

```python
x = ['mon', 'tues', 'wed', 'thu', 'fri'] #x-axis
y = [10, 15, 7, 20, 12] #y-axis

plt.plot(x,y)
plt.title('Weekly Sale')
plt.xlabel('Days of week')
plt.ylabel('Sales')
plt.show()
```
